# Cleaning the data 

```python
# We drop columns that we wont need 

df.drop(["Displayed comment", "Scheduled at", "month", "year", "Cars"], axis = 1, inplace = True)

#Preview and confirm if changes have been effected 

df
```

#Renaming columns and fixing syntax errors 



```python
# For uniformity 

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('_counter', '').str.replace('1.4', '14')

df
```

# Removing outliers 

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1


df_without_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))]
df_without_outliers.head(5)
```

# Checking for None V

```python
#Checking Null alues 
df_without_outliers.isnull().sum()
```

# Backward filling the data

```python
# Pandas dataframe.bfill() is used to backward fill the missing values in the dataset. It will backward fill the NaN values that are present in the pandas dataframe.

new_filled_data= df_without_outliers.fillna(method='bfill')

new_filled_data.head(3)
```

#Answering business questions 

```python
 # Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
new_filled_data['bluecar_diff'] = new_filled_data.bluecar.diff()
new_filled_data['utilib_diff'] = new_filled_data.utilib.diff()
new_filled_data['utilib_14_diff'] = new_filled_data.utilib_14.diff()


paris = new_filled_data['city'] == 'Paris'
pickup = new_filled_data['bluecar_diff'] < 0

paris_p_p_hours = new_filled_data[paris & pickup].groupby('hour')['hour'].value_counts().sort_values(ascending = False).head(1)
paris_p_p_hours
```

```python
# What is the most popular hour for returning cars?

dropoff =new_filled_data['bluecar_diff'] > 0

new_filled_data[dropoff].groupby('hour')['hour'].value_counts().sort_values(ascending = False).head(1)
```

```python
# What postal code is the most popular for picking up Blue cars?


pickup = new_filled_data['bluecar_diff'] < 0

new_filled_data[pickup].groupby('postal_code')['hour'].count().sort_values(ascending = False).head(1)
```
